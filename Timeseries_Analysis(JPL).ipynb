```python
# Assuming 'df' is your DataFrame containing the charging session data

# Convert 'connectionTime' column to datetime if it's not already in datetime format
df['connectionTime'] = pd.to_datetime(df['connectionTime'])

# Extract the day of the week (0 = Monday, 6 = Sunday)
df['day_of_week'] = df['connectionTime'].dt.dayofweek

# Map day of the week to the day name
day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['day_name'] = df['day_of_week'].apply(lambda x: day_names[x])

# Count the number of charging sessions per day type
session_counts = df.groupby('day_name')['sessionID'].count()

# Plot the counts
plt.figure(figsize=(10, 6))
session_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Charging Sessions by Day')
plt.xlabel('Day')
plt.ylabel('Number of Charging Sessions')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the charging session data

# Convert 'connectionTime' and 'disconnectTime' columns to datetime if they're not already in datetime format
df['connectionTime'] = pd.to_datetime(df['connectionTime'])
df['disconnectTime'] = pd.to_datetime(df['disconnectTime'])

# Calculate the charging time for each session
df['charging_time'] = df['disconnectTime'] - df['connectionTime']

# Extract the date from the 'connectionTime' column
df['date'] = df['connectionTime'].dt.date

# Group the data by both 'userID' and 'date' and calculate summary statistics
user_daily_summary = df.groupby(['userID', 'date']).agg({'charging_time': 'sum'}).reset_index()

# Plotting
plt.figure(figsize=(12, 8))

# Plotting for each user
for user_id, user_data in user_daily_summary.groupby('userID'):
    plt.plot(user_data['date'], user_data['charging_time'].dt.total_seconds() / 3600, label=f'{user_id}', marker='o')

plt.title('Total Charging Time of Individual Customers on a Daily Basis')
plt.xlabel('Date')
plt.ylabel('Total Charging Time (hours)')
plt.legend(title='User ID', loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the charging session data

# Sample around 6000 rows from the DataFrame
df_sample = df.sample(n=50, random_state=42)

# Convert 'connectionTime' and 'disconnectTime' columns to datetime if they're not already in datetime format
df_sample['connectionTime'] = pd.to_datetime(df_sample['connectionTime'])
df_sample['disconnectTime'] = pd.to_datetime(df_sample['disconnectTime'])

# Calculate the charging time for each session
df_sample['charging_time'] = df_sample['disconnectTime'] - df_sample['connectionTime']

# Extract the date from the 'connectionTime' column
df_sample['date'] = df_sample['connectionTime'].dt.to_period('M')  # Group by month

# Group the sampled data by both 'userID' and 'date' and calculate summary statistics
user_monthly_summary = df_sample.groupby(['userID', 'date']).agg({'charging_time': 'sum'}).reset_index()

# Plotting
plt.figure(figsize=(12, 8))

# Plotting for each user
for user_id, user_data in user_monthly_summary.groupby('userID'):
    plt.plot(user_data['date'].dt.to_timestamp(), user_data['charging_time'].dt.total_seconds() / 3600, label=f'{user_id}', marker='o')

plt.title('Total Charging Time of Individual Customers on a Monthly Basis (Sampled)')
plt.xlabel('Date')
plt.ylabel('Total Charging Time (hours)')
plt.legend(title='User ID', loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
