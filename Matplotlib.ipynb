This library is useful for making plots.

```python
# Importing the library
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y= np.sin(x)
z= np.cos(x)
print(x.shape)
```

```python
print(x)
print(y)
print(z)
```

```python
plt.plot(x,y) # sine wave
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

```python
# Adding title x-axis and y-axis labels
plt.plot(x,y)
# plt.xlabel('angle')
# plt.ylabel('sine value')
# plt.title('sine wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10,10,20)
y=x**2
print(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
x = np.linspace(-5,5,50)
```

```python
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--') # Implementing multiple functions in one plot
plt.show()
```

```python
#Bar plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
 # here 0,0 is for origin and 1,1 is for height and width
languages = ['English','Telugu','Hindi','French','German']
people = [100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('People')
plt.show()
```

```python
#Pie chart
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','Telugu','Hindi','French','German']
people = [100,50,150,40,70]
ax.pie(people,labels=languages,autopct='%1.1f%%') # autopct gives the floating point numbers
plt.show()
```

```python
# Scatterplot - > very usefulin clustering applications
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig3 = plt.figure()
ax = fig3.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
# 3D scatter plot
fig4 = plt.figure()
ax = plt.axes(projection='3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
